You are serving as a Business Intelligence Analyst at the headquarter of an international fashion goods chain store. Your boss today asked you to do two things for her:

**First, identify two groups of customers from the data set.** The first group is **VIP Customers** whose **aggregated expenses** at your global chain stores are **above the 95th percentile** (aka. 0.95 quantile). The second group is **Preferred Customers** whose **aggregated expenses** are **between the 75th and 95th percentile**.

## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

```python
# import required libraries
import numpy as np
import pandas as pd
```

Next, extract and import `Orders` dataset into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
orders = pd.read_csv('orders.zip')

orders.head()
```

---

"Identify VIP and Preferred Customers" is the non-technical goal of your boss. You need to translate that goal into technical languages that data analysts use:

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub problem 1:

orders_spent = pd.DataFrame(orders.groupby(['CustomerID'])['amount_spent'].sum())

orders_spent.sort_values(['CustomerID'], ascending = True)

```

```python
# Sub problem 2

percent_75 = round(orders_spent['amount_spent'].quantile(0.75),2)

percent_95 = round(orders_spent['amount_spent'].quantile(0.95),2)

print(percent_75, percent_95)
```

```python
# Sub problem 3

preferred_customer = orders_spent[orders_spent['amount_spent'].between(percent_75, percent_95)]

vip_customer = orders_spent[orders_spent['amount_spent'] >= percent_95]

display(vip_customer.sort_values(['amount_spent'], ascending = True))

display(preferred_customer.sort_values(['amount_spent'], ascending = True))
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
orders['VIP Customer'] = np.where(orders['CustomerID'].isin(vip_customer.index), 'VIP', 'Not VIP')

vip = orders[orders['VIP Customer'] == 'VIP']

countries_with_vip = vip[['VIP Customer', 'Country']].groupby(['Country']).count()

display(countries_with_vip.sort_values('VIP Customer'))
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
orders['Preferred Customer'] = np.where(orders['CustomerID'].isin(preferred_customer.index), 'Preferred Customer', 'Not Preferred')

vip_and_preferred = orders[(orders['VIP Customer'] == 'VIP') | (orders['Preferred Customer'] == 'Preferred Customer')]

vip_preferred_countries = pd.DataFrame(vip_and_preferred.groupby(['Country'])['CustomerID'].count())

display(vip_preferred_countries.sort_values('CustomerID', ascending = False))
```
